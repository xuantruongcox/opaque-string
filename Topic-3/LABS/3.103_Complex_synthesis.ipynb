# Exploring complex numbers and complex synthesis

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Properties of sinusoids: phase, frequency and amplitude

```python
fig,ax = plt.subplots(ncols=2, nrows=3, figsize=(16,9), sharex=True, sharey=True)

phase = np.linspace(0, np.pi*2, 100)

## different amplitude
amplitude = 0.5 # feel free to change this
signal = np.sin(phase)
signal2 = np.sin(phase) * amplitude 

ax[0][0].plot(signal)
ax[0][0].set_title("Amplitude 1")

ax[0][1].plot(signal2)
ax[0][1].set_title("Amplitude " + str(amplitude))

## different frequency

frequency = 10 # edit if you want
signal = np.sin(phase)
signal2 = np.sin(phase * frequency) # 

ax[1][0].plot(signal)
ax[1][0].set_title("Frequency 1")

ax[1][1].plot(signal2)
ax[1][1].set_title("Frequency " + str(frequency))

## phase offset
signal = np.sin(phase)
signal2 = np.sin(phase + np.pi/4)

ax[2][0].plot(signal)
ax[2][0].set_title("Phase offset 0")

ax[2][1].plot(signal2)
ax[2][1].set_title("Phase offset pi / 4")


```

## Reminder: simple synthesis

```python
# adapted from thinkdsp p65
def synthesize_simple(amps, fs, ts):
    args = np.outer(ts, fs) # 
    M = np.cos(np.pi*2 * args) 
    ys = np.dot(M, amps)
    return ys

amps = np.array([0.5, 0.25, 0.125])
freqs = np.array([1, 2, 3])
times = np.linspace(0, 1, 100)
signal = synthesize_simple(amps, freqs, times)

plt.plot(signal)
plt.title("Output of simple synthesis")
```

## Complex numbers

```python
# create a complex number
c = 1j * 0.5 # note the 'j' which is the imaginary unit, 1j is 'i'
print(type(c))
print(c.real)
print(c.imag)
```

## The np.exp function

```python
phase = np.linspace(0, np.pi * 4, 100)
npexp = np.exp(phase)
fig, ax = plt.subplots(ncols=3, figsize=(10,3), sharex=True)

ax[0].plot(phase, np.exp(phase))
ax[0].set_title("np.exp(phase)")

ax[1].plot(phase, np.exp(1j * phase).real) # type: ignore
ax[1].set_title("np.exp(1j * phase) real")

ax[2].plot(phase, np.exp(1j * phase).imag) # type: ignore
ax[2].set_title("np.exp(1j * phase) imag")
```

## Complex synthesis

```python
# from thinkdsp p80
def synthesize_complex(amps, fs, ts):
    args = np.outer(ts, fs)
    M = np.exp(1j*(np.pi*2 * args)) # 1j converts to complex 
    print("M contains...", type(M[0][0]))
    ys = np.dot(M, amps)
    return ys

    
amps = np.array([0.5, 0.5, 0.5])
freqs = np.array([1, 2, 3])
times = np.linspace(0, 1, 100)
signal = synthesize_complex(amps, freqs, times)

# notice we pull out real and imaginary parts
plt.plot(signal.real, linestyle="dashed", label="real")
plt.plot(signal.imag, label="imag")

plt.legend(loc="lower left")
plt.title("Output of complex sytnthesis")
        

```
