```python
sig = [1.0, 0.75, 0.5, 0.75, 1.0]
sys = [0.0, 1.0, 0.75, 0.5, 0.25]
```

```python
import numpy as np
def convolve(sig, sys):
  res = np.zeros(len(sig) + len(sys) - 1)
  for sample_ind in range(len(sig)):
    sample = sig[sample_ind]
    for coeff_ind in range(len(sys)):
      output = sample*sys[coeff_ind]
      res_ind = sample_ind + coeff_ind
      res[res_ind] += output
      
  return res
convolved = convolve(sig, sys)
print(convolved)
```
